```python
#removing rows with null values
df = df.dropna(how='all',axis=0) 
```

```python
#"product" in column product is not specified so we will delete rows that have product
df = df.query("Product != 'Product'")
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])#converting to standard date time format
df['Quantity Ordered'] = df['Quantity Ordered'].astype(int)#converting to int data type
df['Price Each'] = df['Price Each'].astype(float)#converting to float data type
df['Order ID'] = df['Order ID'].astype(int)#converting to int data type
```

```python
df['total_price'] = df['Price Each'] * df['Quantity Ordered']
```

```python
df['year'] = df['Order Date'].dt.year
df['month'] = df['Order Date'].dt.month
df['day'] = df['Order Date'].dt.day
df['day_of_week'] = df['Order Date'].dt.day_name()
df['hour'] = df['Order Date'].dt.hour
```

```python
#splitting address
df[['street', 'city', 'state']] = df['Purchase Address'].str.split(',',expand  = True)
```

```python
#it will remove the "street" column from the DataFrame
del df['street']
```

```python
df['state'] = df.state.str[:3]
```

```python
df.reset_index(inplace=True)
del df['index']
```

```python
#making a new column called product type
df['Product Type'] = df.Product.apply(lambda x : x.split()[-1])
df['Product Type'].replace('(4-pack)','Batteries',inplace= True)
df['Product Type'].replace('iPhone','Phone',inplace= True)
```

```python
def sinplot(flip=1): 

    x = np.linspace(0, 15, 100) 

    for i in range(1, 5): 

        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip) 


sns.set_style("white")

sns.set_palette("husl") 

sinplot() 

plt.show()

```
